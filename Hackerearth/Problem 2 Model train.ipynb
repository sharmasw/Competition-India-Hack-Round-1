```python
import pandas as pd
import numpy as np
import xgboost as xgb
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
path='C:/Projects/Hackerearth/Problem 2/'
# path=''
from sklearn import cross_validation, metrics   #Additional scklearn functions
from sklearn.grid_search import GridSearchCV 
from sklearn import metrics,ensemble,linear_model
```

```python
data=pd.read_csv(path+'new_trains.txt')
data.shape
```

```python
data_copy=data.copy()
target=data['segment']
del data
```

```python
del data_copy['segment']
```

```python
X_train,X_test,y_train,y_test=cross_validation.train_test_split(data_copy,target,test_size=.3,stratify=target)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
train_batch=xgb.DMatrix( X_train, label=y_train)
val_batch=xgb.DMatrix(X_test,label=y_test)
```

```python
np.unique(target,return_counts=True)
```

```python
num_rounds=1000
```

```python
param = {'max_depth':2, 'eta':1, 'silent':1, 'objective':'binary:logistic','eval_metric':'auc',
        'learning_rate' :0.01, 'n_estimators':140, 'max_depth':6,
 'min_child_weight':1, 'gamma':0, 'subsample':0.8, 'colsample_bytree':0.8,
 'objective': 'binary:logistic', 'nthread':4, 'scale_pos_weight':1,'seed':27}
wathclist=[(val_batch,'eval'), (train_batch,'train')]
model = xgb.train(param, train_batch, num_rounds,wathclist,early_stopping_rounds=20) 
```

```python
pred=model.predict(val_batch)
metrics.roc_auc_score(y_test,pred)
```

```python
GBR_model=ensemble.GradientBoostingClassifier().fit(X_train,y_train)
metrics.roc_auc_score(y_test,GBR_model.predict_proba(X_test)[::,1])
```

```python
Line_model=linear_model.LinearRegression().fit(X_train,y_train)
metrics.roc_auc_score(y_test,Line_model.predict(X_test))
```

```python
pred_GBR_model=GBR_model.predict(X_test)
pred_Line_model=Line_model.predict(X_test)
```
